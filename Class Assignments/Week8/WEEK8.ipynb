```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python

gdp = pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_3052552.csv', skiprows=4)
jrn = pd.read_csv('API_IP.JRN.ARTC.SC_DS2_en_csv_v2_3057620.csv', skiprows=4)
un_countries =  pd.read_csv("UNSD — Methodology.csv")
```

```python
codes  = un_countries["ISO-alpha3 Code"].dropna()
gdp_countries = gdp[gdp["Country Code"].isin(codes)]
jrn_countries = jrn[jrn["Country Code"].isin(codes)]
gdp_countries.set_index("Country Code", inplace=True)
jrn_countries.set_index("Country Code", inplace=True)
```
